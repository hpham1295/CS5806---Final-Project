```python
import pandas as pd
import os
import matplotlib.pyplot as plt

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
# --- Step 0: Load the data ---'
save_path = r'C:/Users/anhuy/CS-5806/Project/data/processed'  # Make sure this matches

X = pd.read_csv(os.path.join(save_path, 'X_subsampled_trimmed.csv'))
y = pd.read_csv(os.path.join(save_path, 'y_subsampled_trimmed.csv')).squeeze()

lat_lon = pd.read_csv(os.path.join(save_path, 'lat_lon_subsampled_trimmed.csv')).squeeze()
original_index = pd.read_csv(os.path.join(save_path, 'original_index_subsampled_trimmed.csv')).squeeze()

# First split: 70% train, 30% validation + test
X_train, X_temp, y_train, y_temp, latlon_train, latlon_temp, idx_train, idx_temp = train_test_split(
    X, y, lat_lon, original_index, test_size=0.3, random_state=42, shuffle=True)

# Second split: 50% validation, 50% test on the 30% validation + split from the first split (i.e, 15% validation and 15% test in the entire dataset)
X_val, X_test, y_val, y_test, latlon_val, latlon_test, idx_val, idx_test = train_test_split(
    X_temp, y_temp, latlon_temp, idx_temp, test_size=0.5, random_state=42, shuffle=True)
```

```python
# --- Step 1: Define the parameter grid ---
param_grid = {
    'max_depth': [5, 10, 15],
    'n_estimators': [100, 200, 300],
    'max_features': ['sqrt', 'log2']
}

# --- Step 2: Initialize the model (base model) ---
model = RandomForestRegressor(
    random_state=42
)

# --- Step 3: Set up GridSearchCV ---
grid_search = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    scoring='neg_mean_squared_error', 
    cv=3,
    verbose=2,
    n_jobs=-1
)

# --- Step 4: Run the search ---
grid_search.fit(X_train, y_train)

# --- Step 5: Get the best model and parameters ---
print(f"Best Parameters: {grid_search.best_params_}")
print(f"Best Validation Score (Negative MSE): {grid_search.best_score_:.4f}")

best_model = grid_search.best_estimator_

# --- Step 6: Evaluate on Validation and Test set ---
y_val_pred = best_model.predict(X_val)
```

```python
# Extract GridSearchCV results
results = pd.DataFrame(grid_search.cv_results_)

# Plot test scores vs max_depth
plt.figure(figsize=(4,5))
for mf in param_grid['max_features']:
    subset = results[results['param_max_features'] == mf]
    plt.plot(subset['param_max_depth'], -subset['mean_test_score'], marker='o', label=f'max_features={mf}')

plt.xlabel('Max Depth')
plt.ylabel('Validation MSE')
# plt.title('Validation MSE vs Max Depth for Different Max Features')
plt.legend()
plt.grid(True)
plt.tight_layout()

# plt.savefig('C:/Users/anhuy/CS-5806/Project/photo/RF_1.png',dpi=300, bbox_inches='tight')

plt.show()
```

```python
# NOTE: Random Forest does not have `staged_predict` like Boosting.
# We'll simulate by training with different number of trees.

train_loss = []
val_loss = []
n_estimators_list = list(range(10, best_model.n_estimators + 10, 10))  # like [10,20,30,...]

for n in n_estimators_list:
    temp_model = RandomForestRegressor(
        n_estimators=n,
        max_depth=best_model.max_depth,
        max_features=best_model.max_features,
        random_state=42
    )
    temp_model.fit(X_train, y_train)
    
    # Training Loss
    y_pred_train = temp_model.predict(X_train)
    train_loss.append(mean_squared_error(y_train, y_pred_train))
    
    # Validation Loss
    y_pred_val = temp_model.predict(X_val)
    val_loss.append(mean_squared_error(y_val, y_pred_val))
```

```python
# Plot Training and Validation Loss vs Number of Trees
plt.figure(figsize=(4,5))
plt.plot(n_estimators_list, train_loss, label='Training Loss')
plt.plot(n_estimators_list, val_loss, label='Validation Loss')
plt.xlabel('Number of Trees (Estimators)')
plt.ylabel('Mean Squared Error (MSE)')
# plt.title('Training and Validation Loss vs Number of Trees')
plt.legend()
plt.grid(True)
plt.tight_layout()

# plt.savefig('C:/Users/anhuy/CS-5806/Project/photo/RF_3.png',dpi=300, bbox_inches='tight')

plt.show()
```

```python
# From the plots above, the best number of trees to prevent overfitting is 100 trees. The best max feature and tree depth are sqrt and 15
best_model_100 = RandomForestRegressor(
    max_features='sqrt',
    max_depth=15,
    n_estimators=100,
    random_state=42
)
best_model_100.fit(X_train, y_train)
```

```python
# Predict on the test set
y_test_pred = best_model_100.predict(X_test)

# Evaluate performance on test set
print(f"Test MSE: {mean_squared_error(y_test, y_test_pred):.4f}")
print(f"Test R2: {r2_score(y_test, y_test_pred):.4f}")

# Plot True vs. Predicted
plt.figure(figsize=(4,5))
plt.scatter(y_test, y_test_pred, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel("True UHII")
plt.ylabel("Predicted UHII")
# plt.title("True vs Predicted UHII")
plt.grid(True)

# plt.savefig('C:/Users/anhuy/CS-5806/Project/photo/RF_2.png',dpi=300, bbox_inches='tight')

plt.show()
```
